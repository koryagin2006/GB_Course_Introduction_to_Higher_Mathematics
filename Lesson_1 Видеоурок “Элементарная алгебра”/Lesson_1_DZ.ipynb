### 2.2 Задание (на листочке) - Решите систему уравнений:

* x<sup>2</sup> + y∙x – 9 = 0
* x – y/5 = 0

Линейная или нелинейная это система? 
А каждое уравнение по отдельности?

**Решение:**

Система нелинейная. 1 уравнение квадратное, 2 - линейное
- y = 5\*x
- x<sup>2</sup> + 5x<sup>2</sup> - 9 = 0
- 6x<sup>2</sup> - 9 = 0
-  
- 6x<sup>2</sup> +0x + (-9) = 0

````python
# Находим дискриминант
a = 6
b = 0
с = -9
D = b**2 - 4*a*с
print(D>0)
# 2 корня
````

````python
x1 = (-b + sqrt(D)) / (2*a)
x2 = (-b - sqrt(D)) / (2*a)

y1 = 5*x1
y2 = 5*x2

print('x1=',x1)
print('y1=',y1)

print('-'*30)

print('x2=',x2)
print('y2=',y2)
````

## 3. Задание (на листочке) - Решите задачу:
Площадь пола прямоугольной комнаты равна 48 м2,а его периметр равен 28 м. Найдите длину и ширину комнаты.
```
x+y=28
x*y=48
```

Решение:
- x*(28-x)=48
- 28x-x<sup>2</sup> = 48
-  
- -1x<sup>2</sup>+28x-48=0

````python
# Находим дискриминант
a = -1
b = 28
c = -48
D = b**2 - 4*a*c

print(D>0)
# 2 корня
````

````python
x = (-b + sqrt(D)) / (2*a)
print('x=',x)
print('y=',28-x)
````

````python
x = (-b - sqrt(D)) / (2*a)
print('x=',x)
print('y=',28-x)
````

## 4. Задание (в программе):
Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

````python
x = np.linspace(-10,10,300)

def y(k):
    return np.cos(k*x)

plt.figure(figsize=(16,4))

plt.plot(x, y(1), label='k=1')
plt.plot(x, y(3), label='k=3')

plt.title('График cos(x)')
plt.xlabel('Переменная x')
plt.ylabel('Функция f(x)')
plt.legend()
plt.show()
````
