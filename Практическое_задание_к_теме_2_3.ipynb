## 3. Задание (в программе)

Напишите код на Python, реализующий построение графиков:

### - окружности,

```
(x-x0)^2 + (y-y0)^2 = R^2

если x0 и y0 == 0, то 
x^2 + y^2 = R^2

y = sqrt(R**2 - x**2)
```

````python
plt.figure(figsize=[5,5])
R=5
x = np.linspace(-5, 5, 200)
y1 = np.sqrt(R**2 - x**2)
y2 = -np.sqrt(R**2 - x**2)

plt.plot(x,y1)
plt.plot(x,y2)

plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-6,6)
plt.ylim(-6,6)
plt.title('Окружность')
plt.grid();
````

## 1. Задание (в программе)
Нарисуйте график функции:
```
y(x) = k∙cos(x – a) + b
```
для некоторых (2-3 различных) значений параметров k, a, b

````python
def cos_(k,a,b):
  x = np.linspace(-2*np.pi, 3*np.pi, 201) 
  y = k*np.cos((x-a)) + b
  plt.plot(x, y, label='{}*np.cos((x-{})) + {}'.format(k,a,b))
  plt.xlabel('x')
  plt.ylabel('y')
  plt.plot([-7,12],[0,0], color ='black', linewidth=1, linestyle ="--")
  plt.plot([0,0],[-6,12], color ='black', linewidth=1, linestyle ="--")
  plt.legend()
  plt.grid(True)
  plt.title('Графики косинуса');
````

````python
R = int(input('Введите длину R: '))
alpha = int(input('Введите угол в грудусах: '))

alpha = np.radians(alpha)
x = round(R*np.cos(alpha),1)
y = round(R*np.sin(alpha),1)

print('координаты точки: ({}, {})'.format(x,y))
````

#### - Напишите код, который будет рисовать график окружности в полярных координатах.

````python
def circle(R,alpha):
  plt.figure(figsize=[5,5])
  x = R*np.cos(alpha)
  y = R*np.sin(alpha)
  
  plt.plot(x,y)
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title('Окружность в полярных координатах')
  plt.grid();

circle(R=5, alpha=np.linspace(0, 2*np.pi, 500))
````

#### - Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.

````python
def line_segment(R1,alpha1,R2,alpha2):
  x1 = R1*np.cos(alpha1)
  y1 = R1*np.sin(alpha1)
  x2 = R2*np.cos(alpha2)
  y2 = R2*np.sin(alpha2)

  plt.plot([x1,x2],[y1,y2])
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title('Отрезок в полярных координатах')
  plt.grid();

line_segment(R1 = -3, alpha1 = np.pi/3,
     R2 = 5, alpha2 = 2*np.pi/3)
````

````python
from scipy.optimize import fsolve
````

#### Решите систему уравнений:
- y = x<sup>2</sup> - 1
- exp(x) + x∙(1 - y) = 1

##### Решение
 - x<sup>2</sup> - 1 - y = 0
 - exp(x) + x∙(1 - y) - 1 = 0

````python
x = np.linspace(-10,10, 201)
y1 = x**2 - 1
y2 = (np.exp(x)+x-1)/x
plt.plot(x,y1)
plt.plot(x,y2)
plt.ylim(-2,25)
plt.xlim(-3,6)
plt.minorticks_on()
plt.grid(which='major')
plt.grid(which='minor',linestyle = ':')
plt.title('Решение системы в точках персечения графиков');
````

````python
def equations(p):
  x, y = p
  return (x**2 - 1-y,
          np.exp(x) - x*(1 - y)-1)

for i in fsolve(equations, (-2,2)):
  print('x =',i)
````

#### Решите систему уравнений и неравенств:
- y = x<sup>2</sup> - 1
- exp(x) + x∙(1 – y) > 1

##### Решение:
- y = x<sup>2</sup> - 1
- y < (exp(x) + x - 1)/x

````python
x = np.linspace(-10,10, 201)
y1 = x**2 - 1
y2 = (np.exp(x)+x-1)/x

plt.plot(x,y1)
plt.plot(x,y2)

plt.ylim(-5,25)
plt.xlim(-5,6)
plt.minorticks_on()
plt.grid(which='major')
plt.grid(which='minor',linestyle = ':')
plt.title('Решение системы в точках персечения графиков');
````
