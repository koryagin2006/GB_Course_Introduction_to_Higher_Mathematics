Напишите код на Python, реализующий построение графиков:

### - окружности,

```
(x-x0)^2 + (y-y0)^2 = R^2
  если x0 и y0 == 0, то 
  x^2 + y^2 = R^2
y = sqrt(R**2 - x**2)
```

````python
plt.figure(figsize=[5,5])
R=5
x = np.linspace(-5, 5, 200)
y1 = np.sqrt(R**2 - x**2)
y2 = -np.sqrt(R**2 - x**2)

plt.plot(x,y1)
plt.plot(x,y2)

plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-6,6)
plt.ylim(-6,6)
plt.title('Окружность')
plt.grid();
````

### - эллипса,

````python
plt.figure(figsize=[5,5])
a=2
b=10
x = np.linspace(-50, 50, 5000)
y1 = np.sqrt(b**2 - (x**2-b**2)/(a**2))
y2 = -np.sqrt(b**2 - (x**2-b**2)/(a**2))

plt.plot(x,y1)
plt.plot(x,y2)
plt.xlim(-50,50)
plt.ylim(-50,50)
plt.xlabel("x")
plt.ylabel("y")
plt.title('Эллипс')
plt.grid();
````

## 1). Задание **[сделано]** (в программе)
Нарисуйте график функции:
```
y(x) = k∙cos(x – a) + b
```
для некоторых (2-3 различных) значений параметров k, a, b

````python
rcParams['figure.figsize'] = 10, 4

def cos_(k,a,b):
  x = np.linspace(-2*np.pi, 3*np.pi, 201) 
  y = k*np.cos((x-a)) + b
  plt.plot(x, y, label='{}*np.cos((x-{})) + {}'.format(k,a,b))
  plt.xlabel('x')
  plt.ylabel('y')
  plt.plot([-7,12],[0,0], color ='black', linewidth=1, linestyle ="--")
  plt.plot([0,0],[-6,12], color ='black', linewidth=1, linestyle ="--")
  plt.legend()
  plt.grid(True)
  plt.title('Графики косинуса');
````

#### 3.2)
- Напишите код, который будет рисовать график окружности в полярных координатах.

````python
def circle(R,alpha):
  plt.figure(figsize=[5,5])
  x = R*np.cos(alpha)
  y = R*np.sin(alpha)
  
  plt.plot(x,y)
  plt.xlabel("x")
  plt.ylabel("y")
  plt.title('Окружность в полярных координатах')
  plt.grid();

circle(R=5, alpha=np.linspace(0, 2*np.pi, 500))
````

````python
# Вариант 2.
def xy2pol(x,y):
  """ Перевод в полярные координаты
  """
  rho = np.sqrt(x**2 + y**2)
  phi = np.arctan2(y,x)*180/np.pi
  return (rho,phi)

t = np.linspace(0, np.pi, 1000)
x = np.cos(t)
y = np.sin(t)
rho, phi = xy2pol(x,y)
plt.polar(phi, rho);
````
