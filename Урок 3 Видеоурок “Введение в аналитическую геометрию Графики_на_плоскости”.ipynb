## Тригонометрические функции

## Примеры графиков

### Полиномы

Полином (многочлен)

y(x) = A<sub>2</sub> \* x<sup>2</sup> + A<sub>1</sub> * x + A<sub>0</sub> - полином 2 степени

y(x) = A<sub>3</sub> \* x<sup>3</sup> + A<sub>2</sub> \* x<sup>2</sup> + A<sub>1</sub> * x + A<sub>0</sub> - полином 3 степени

#### График полинома

```python
x = np.linspace(-1, 3, 201)
print(np.poly([1, 1, 2]))  # выводит коэффициенты для заданных условий (количество корней)
plt.plot(x, 2**x)  # график показательной функции
plt.plot(x,np.polyval([1., -4.,  5., -2.],x))  # вычисляет значение коэфициентов полинома в определенной точке x. графики одинаковые
# plt.plot(x, np.exp(x))
plt.scatter(1,0,color='red', label='локальный максимум')
# plt.scatter(1,0,color='red', label='локальный максимум')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-1,3)
plt.ylim(-1,3)
plt.grid(True)
plt.legend()
plt.show()
print(np.roots([ 1., -4.,  5., -2.]))  # считает корни уравнения (в параметры передаются коэффициенты уравнения)
```

## Преобразование координат (примеры)

Нарисуем 2 параболы

```python
def parabols(a,b):
  x = np.linspace(-20, 20, 1000)
  y = (x-a)**2+b
  plt.plot(x, y, label='(x-{})^2 + {}'.format(a,b))
  plt.xlabel('x')
  plt.ylabel('y')
  plt.legend()
  plt.xlim(-10,30)
  plt.ylim(-30,30)
  plt.title('Параболы')
  plt.grid(True);

parabols(a=0,b=0)
parabols(a=5,b=-20)

# мы совершили пераллельный перенос параболы
```

## Графическое и численное решение уравнений

Численное решение уравнений

sin(x+1) = 0
переводим:

y = sin(x+1)
 и смотрим, при каких x график пересекает 0

```python
from scipy.optimize import fsolve
```

```python
def f(x):
  return (np.sin(x+1))

a = 5  # наименьшее приближение для поиска корня функции
x0 = fsolve(f, a)
print('решение: ',x0)

x = np.linspace(-3*np.pi, 3*np.pi, 201)
plt.plot(x, f(x))
plt.plot([-10,10],[0,0],linestyle ="--")
plt.plot([a,a],[-1,1])

plt.grid(True);
```

Решение системы уравнений:
- y = x<sup>2</sup> - 1
- exp(x) - xy = 1

Переводим части уравнений:
- y- x<sup>2</sup> + 1 = 0
- exp(x) - xy - 1 = 0
-------------------

- y(x) = x<sup>2</sup> - 1
- y(x) = (exp(x) -1)/x

```python
x = np.linspace(-2,3, 201)
y1 = x**2 - 1
y2 = (np.exp(x) - 1)/x
plt.plot(x,y1)
plt.plot(x,y2)
plt.ylim(-1,5)
plt.minorticks_on()
plt.grid(which='major')
plt.grid(which='minor',linestyle = ':')
plt.title('Решение системы в точках персечения графиков');
```
